```python
import pandas as pd
```

Merge menu với order_log theo product_id

```python
df_order = pd.read_csv('order_log_clean.csv')
df_menu = pd.read_csv('menu_clean.csv')
df_merged = pd.merge(df_order, df_menu, on='product_id', how='left')
print(f"Số dòng: {df_merged.shape[0]}")
print(f"Số cột: {df_merged.shape[1]}")
print(f"Các cột: {list(df_merged.columns)}")
```

Merge với cost_raw_material để tính lợi nhuận

```python
df_cost = pd.read_csv('cost_raw_material_clean.csv')
df_merged = pd.merge(df_merged, df_cost, on='product_id', how='left')
print(f"Số dòng: {df_merged.shape[0]}")
print(f"Số cột: {df_merged.shape[1]}")
# Tính lợi nhuận
df_merged['revenue'] = df_merged['qty'] * df_merged['price']
df_merged['total_cost'] = df_merged['qty'] * df_merged['unit_cost']
df_merged['profit'] = df_merged['revenue'] - df_merged['total_cost']
```

Phát hiện mismatch hoặc món không có cost

```python
# Kiểm tra product_id không có trong menu
missing_menu = df_merged[df_merged['product_name'].isna()]
if len(missing_menu) > 0:
    print(f"Có {len(missing_menu)} dòng không có trong menu")
    print(f"Product IDs: {missing_menu['product_id'].unique()}")
else:
    print(f"Tất cả product_id đều có trong menu")
# Kiểm tra product_id không có cost
missing_cost = df_merged[df_merged['unit_cost'].isna()]
if len(missing_cost) > 0:
    print(f"Có {len(missing_cost)} dòng không có cost")
    print(f"Product IDs: {missing_cost['product_id'].unique()}")
else:
    print(f"Tất cả product_id đều có cost")
```

Xuất file dữ liệu hoản chỉnh

```python
output_file = 'complete_coffee_shop_data.csv'
df_merged.to_csv(output_file, index=False)
print(f"Đã xuất file: {output_file}")
print(f"Số dòng: {df_merged.shape[0]}")
print(f"Số cột: {df_merged.shape[1]}")
```
