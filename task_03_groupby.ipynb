```python
# TASK_3

import pandas as pd
```

```python
# CHẠY  TASK1
menu = pd.read_csv("menu_clean.csv")
order_log = pd.read_csv("order_log_clean.csv")
cost = pd.read_csv("cost_raw_material_clean.csv")
menu.head()
```

```python
df = pd.merge(order_log, menu, on="product_id", how="left")
   # ghép với menu để lấy thông tin món
df = pd.merge(df, cost, on="product_id", how="left")
   # ghép với cost để lấy thông tin chi phí
print(df.dtypes)
```

```python
# ép kiểu dữ liệu về số để tính toán
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['unit_cost'] = pd.to_numeric(df['unit_cost'], errors='coerce')
df['qty'] = pd.to_numeric(df['qty'], errors='coerce')
```

```python
# tính toán doanh thu và lợi nhuận
df['doanh_thu'] = df['price'] * df['qty']
df['loi_nhuan'] = (df['price'] - df['unit_cost']) * df['qty'] 
loi_nhuan_nhom = df.groupby('product_id')['loi_nhuan'].sum()   # lợi nhuận theo nhóm sản phẩm
```

```python
#Tìm nhóm có biên lợi nhuận cao nhất
nhom_max_loi_nhuan = loi_nhuan_nhom.idxmax()
print("Nhóm sản phẩm có biên lợi nhuận cao nhất:", nhom_max_loi_nhuan)
```

```python
# Tính tổng số lượng bán theo từng món
tong_so_luong = df.groupby('product_id')['qty'].sum()
print(tong_so_luong)
# Tính số đơn hàng chứa từng món
so_don_hang = df['product_id'].value_counts()
print(so_don_hang)
# Hiệu suất = tổng số lượng / số đơn hàng
hieu_suat = tong_so_luong / so_don_hang
print(hieu_suat)
# Tìm món có hiệu suất thấp nhất
mon_hieu_suat_thap = hieu_suat.idxmin()
print("Món có hiệu suất thấp nhất:", mon_hieu_suat_thap)
print("Giá trị hiệu suất:", hieu_suat.min())
```
