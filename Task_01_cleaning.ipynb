### Chuẩn hóa product_id và tên món

**Tổng quan dữ liệu**

```python
menu.dtypes
```

 **chuẩn hóa product_id**

```python
menu["product_id"] = (
    menu["product_id"]
    .astype(str)# đảm bảo dữ liệu là chuỗi 
    .str.strip()#xóa khoảng trắng thừa
)
print(menu.head())
```

 **chuẩn hóa tên món**

```python
menu["product_name"] = (
    menu["product_name"]
    .astype(str)
    .str.strip()
    .str.title()# viết hoa chữ cái đầu mỗi từ
)
```

### Làm sạch price: loại khoảng trắng, text. 


**làm sạch price trong menu.csv**

```python
menu["price"] = ( # price về chuỗi để xử lý
    menu["price"]
    .astype(str)
    .str.replace(r"[^\d.]", "", regex=True)#regex xóa mọi ký tự không phải số hoặc dấu chấm
    .str.strip()
)# \d chữ số từ 0-9,. số thập phân,[^ ] tất cae ký tự không thuộc nhóm
print(menu.head(15))
```

### Làm sạch order_time, đưa về định dạng thống nhất. 

```python
order_log["order_time"] = pd.to_datetime(
    order_log["order_time"],
    errors="coerce"
)
print(order_log.head())
```

### Xử lý qty âm hoặc sai định dạng. 

```python
#Xóa các ký tự ''x'' ở dữ liệu qty để tránh lỗi định dạng chuyển về NaN
order_log["qty"] = (
    order_log["qty"]
    .astype(str)
    .str.extract(r"(-?\d+\.?\d*)")[0]   # trích số
    .astype(float)
)

# Lấy trị tuyệt đối đề phòng có dữ liệu âm
order_log["qty"] = order_log["qty"].abs()
# Kiểm tra lại kết quả

order_log.head(10)
```

### Làm sạch unit_cost. 

```python
cost["unit_cost"] = (
    cost["unit_cost"]
    .astype(str)
    .str.replace(r"[^\d.]", "", regex=True)
)
cost["unit_cost"] = pd.to_numeric(cost["unit_cost"], errors="coerce")
```

###  Xuất 3 file sạch để dùng cho các nhiệm vụ sau. 

```python
menu.to_csv("menu_clean.csv", index=False)
order_log.to_csv("order_log_clean.csv", index=False)
cost.to_csv("cost_raw_material_clean.csv", index=False)
```
